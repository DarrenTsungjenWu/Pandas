```python
import pandas as pd
import numpy as np
```

```python
# Create dataframe for reshape example
df = pd.DataFrame({'Day':['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
                   'Chicago':[38, 42, 32, 29, 27, 26, 33],
                   'New York':[36, 32, 34, 28, 29, 30, 29],
                   'Anaheim':[40, 43, 36, 35, 36, 40, 36]
})
df
```

```python
#Using pd.melt() to transform the wide dataframe into long dataframe.
df_trans = df.melt(id_vars = ['Day']) #id_vars is to indicate the columns you wanna keep unchanged.
print(df_trans)
print('See only the result of New York below')
print(df_trans[df_trans['variable'] == 'New York'])
```

# Change the name of variable and value when pd.melt() dataframe

```python
# Change the name of variable and value that are just transformed out from the original data
df_trans_rename = df.melt(id_vars = ['Day'], var_name = "City", value_name = 'Temperature')
df_trans_rename
```

```python
# We can also melt and show the result table with Chicago City only, by defining argment value_vars.
df_test = df.melt(id_vars = ['Day'], value_vars = 'Chicago', var_name = 'Chicago City', value_name = 'Temperature')
df_test
```
